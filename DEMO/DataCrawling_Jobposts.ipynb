## Crawling job post from recruitment website using Selenium and BeautifulSoup

```python
# import library
import time
import requests
import pandas as pd
from selenium import webdriver
from bs4 import BeautifulSoup
driver = webdriver.Chrome('/usr/local/bin/chromedriver')
```

```python
def getjobposts(searchword,numpage):
    # search_word = 'Data Science' # job title
    search_word_ = searchword.replace(' ', '-')
    sw_ = searchword.replace(' ', '_')
    startpage = 1
    jp_list = []
    webname = 'https://hk.jobsdb.com'
    for k in range(startpage,numpage+1):
        url = 'https://hk.jobsdb.com/hk/search-jobs/{}/{}'.format(search_word_,k)
        driver.get(url)
        bs = BeautifulSoup(driver.page_source, 'lxml')

        for i in bs.find_all('div',{'class':'FYwKg _1A6vC_6 _31UWZ _1gtjJ _2cWXo _1Swh0'}):
            titlename = i.find('div',{'class':'FYwKg _2j8fZ_6 sIMFL_6 _1JtWu_6'}).text.strip()
            companyname_ = i.find('span',{'class':'FYwKg _1GAuD C6ZIU_6 _1_nER_6 _27Shq_6 _29m7__6'})
            if companyname_ is not None:
                companyname = companyname_.text.strip()
            if companyname_ is None:
                continue
            posttime = i.find('time',{'class':'FYwKg _1GAuD'})['datetime']
            jobfunc = i.find('div',{'class':'FYwKg _3TDN__6'}).text.strip()
            link = i.find('h1',{'class':'FYwKg _1GAuD C6ZIU_6 _6ufcS_6 _27Shq_6 sQuda_6'}).find('a')['href']
            weblink = webname+link
            jp_list.append({'Title':titlename,'Company':companyname,'Post_Date':posttime,\
                            'Job_Functions':jobfunc,'Link':weblink})

    print('Total number of job posts:',len(jp_list))
    df = pd.DataFrame(jp_list)
    df.to_csv('{}_job_posts.csv'.format(sw_),index=False)
    driver.close()
    return df
```

```python
# E.g. Search : "Data Science", Total number of Pages : 20
getjobposts("Data Science",20)
```
